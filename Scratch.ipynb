```python
import matplotlib.pyplot as plt
```

```python
from jupyterthemes import jtplot
jtplot.style(theme='monokai', context='notebook', ticks=False, grid=False)
```

```python
def plot_zp(zeros, poles, save='', orderzeros={}, orderpoles={}):
    plt.figure()

    plt.axhline(0, color='red')  # x = 0
    plt.axvline(0, color='red')  # y = 0
    for zero in zeros:
        plt.scatter(zero.real, zero.imag, marker='o',
                    edgecolors='w', color='k', linewidths=2, zorder=4)

    for pole in poles:
        plt.scatter(pole.real, pole.imag, marker='x', color='w', zorder=4)

    for zero, order in orderzeros.items():
        for o in range(order, 0, -1):
            plt.scatter(zero.real, zero.imag, marker='o', edgecolors='w',
                        color='k', linewidths=1.5, zorder=4, s=o*120)

    plt.xlabel('Real')
    plt.ylabel('Imaginary')
    plt.savefig(save)
    plt.show()
```

```python
plot_zp([1/2*1j*(95**.5+9j), -1/2*1j*(95**.5-9j)], [-3, -2+5j, -2+5j], './1a.png')
```

```python
plot_zp([2.5], [1,4], './1b.png')
```

```python
plot_zp([-3,3], [0], './1c.png')#, {-3:2,3:2})
```

```python
plot_zp([], [3,-1], './4a.png')#, {-3:2,3:2})
```
